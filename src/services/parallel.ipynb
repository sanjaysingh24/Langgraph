```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
        runs:int
        balls:int
        fours:int
        sixes:int

        sr:float
        bpb:float
        boundary_percent:float
        summary:str
```

```python
def calculateStrikerate(state:BatsmanState)->BatsmanState:
    sr = (state["runs"]/state["balls"])/100
    
    return {'sr':sr}
```

```python
def calculatebpb(state:BatsmanState)->BatsmanState:
    bpb = state['balls']/(state['fours']+state['sixes'])
    
    return {'bpb':bpb}
```

```python
def calculate_boundry(state:BatsmanState):
    boundry_percent=(((state['fours'] *4 )+ (state['sixes']*6))/state['runs'])*100

    return  {'boundary_percent':boundry_percent}
```

```python
def summary(state:BatsmanState):
    summary=f"""
    Strike-Rate - {state['sr']} \n
    balles per boundary - {state['bpb']} \n
    boundary percentage - {state['boundary_percent']}
    """
    
    return {'summary':summary}
```

```python
graph =StateGraph(BatsmanState)

graph.add_node("calculateStrikeRate",calculateStrikerate)
graph.add_node("calculate_bpb",calculatebpb)
graph.add_node("calculate_boundary",calculate_boundry)
graph.add_node("summary",summary)
```

```python
graph.add_edge(START,"calculateStrikeRate")
graph.add_edge(START,"calculate_bpb")
graph.add_edge(START,"calculate_boundary")
graph.add_edge("calculateStrikeRate","summary")
graph.add_edge("calculate_bpb","summary")
graph.add_edge("calculate_boundary","summary")
graph.add_edge("summary",END)

#compile the graph
workflow = graph.compile()

intial_state={
    'runs':100,
    'balls':50,
    'fours':6,
    'sixes':4

}

final_output = workflow.invoke(intial_state)
final_output
```
